```python
plt.figure(figsize=(15, 6))
frequency_table = df['n1'].value_counts().reset_index()
frequency_table.columns = ['n1', 'Frequency']
total_count = frequency_table['Frequency'].sum()
frequency_table['Percentage'] = (frequency_table['Frequency'] / total_count) * 100

# Plot the frequency table
ax = sns.barplot(data=frequency_table, x='n1', y='Frequency')
plt.xlabel('n1')
plt.ylabel('Frequency')
plt.title('Frequency Table for n1')

# Annotate the percentages on the bars
for i, row in frequency_table.iterrows():
    ax.text(i, row['Frequency'], f"{row['Percentage']:.2f}%", ha='center', va='bottom')

plt.xticks(rotation=90)
plt.show()
```

### Categoric Data

```python
i=1
plt.figure(figsize=(20,8))
columns=['c1','c2','c6']
for col in columns:
    top= df[col].value_counts()[:15].index.tolist()
    oth= df[col].value_counts()[15:].sum()
    top.append("Other")
    plot_df = pd.DataFrame({col: df[col]})
    plot_df.loc[~plot_df[col].isin(top), col] = "Other"
    plt.subplot(1,3,i)
    sns.countplot(data=plot_df, x=col,order=top)
    plt.xticks(rotation=90)
    i=i+1
plt.show()
```

```python
top_countries = df['country'].value_counts()[:10].index.tolist()
ax = sns.countplot(data=df, x='country', order=top_countries, hue='b17')
total = len(df)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom')

plt.xticks(rotation=90)
plt.show()
```

we can interpret conversion rates from above countries...
