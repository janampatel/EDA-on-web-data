```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("haberman.csv",header=None,names=["Age","Operation_yr","axil_nodes_det","Sur_det"])
df.head()
```

```python
df.describe()
```

```python
print("This is shape of table: ",df.shape)
```

```python
df.Sur_det.value_counts()
```

```python
sns.distplot(df["Sur_det"])
```

```python
#FacetGrid is used in seaborn library for creating grid-like layouts of plots based on one or more categorical variables
#It allows you to visualize relationships between multiple variables by creating a grid of subplots

#hue parameter is used to specify the categorical variable "Sur_det",
#which means that the grid of plots will be differentiated by different values of "Sur_det".

sns.FacetGrid(df,hue="Sur_det",height=5).map(sns.distplot,"Age").add_legend()
```

from the above graph, you can observe that people age 50 to 60 have more survival rate.

```python
sns.FacetGrid(df,hue="Sur_det",height=5).map(sns.distplot,"Operation_yr").add_legend()
```

from the above graph, we can say that people who had an operation in the year from 58 to 66 had more survival rate.

CDF(cumulative distributive function), PDF(probability density function):

   RV     PDF   CDF
   
(x = 1) : 1/6   1/6

(x = 2) : 1/6   2/6

(x = 3) : 1/6   3/6

(x = 4) : 1/6   4/6

(x = 5) : 1/6   5/6

(x = 6) : 1/6   6/6

```python
counts, bin_edges = np.histogram(df["Age"],bins=10)
print(bin_edges)
pdf= counts/sum(counts)
print("PDF: ",pdf)
cdf=np.cumsum(pdf) #cumsum is used for cummulative sum
print("CDF: ",cdf)

plt.plot(bin_edges[1:],pdf) #bin edges has 11 value, so we skip the first value
plt.plot(bin_edges[1:],cdf)
```
